# Calculation of the Morgan fingerprint

```python

from rdkit.Chem import AllChem
#MFP = AllChem.GetMorganFingerprintAsBitVect(df["SMILES"],5,nBits=512)
print(df['SMILES'])
```

```python
def SmilesToFp(x):
    mol = Chem.MolFromSmiles(x)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 5,nBits=512)
    return np.array(fp)
df["Morgan"] = df.apply(lambda row: SmilesToFp(row['SMILES']), axis=1)
```

```python
df.head()
```

```python
InChIKey = [df['InChIKey']]
headers = ['InChIKey']
df_InChIKey = pd.concat(InChIKey, axis = 1, keys = headers)

df_Morgan = pd.DataFrame(df['Morgan'].tolist())

df2 = pd.concat([df_InChIKey, df_Morgan], axis = 1)
```

```python
df2.head()
```

```python
#export DataFrame to CSV file
df2.to_csv(r'C:\documenten hier\MorganFp_ccr2.csv', index=False)
```

# Calculation of 2D descriptors

```python
from rdkit.Chem import Descriptors
from rdkit.Chem.Descriptors import ExactMolWt
```

```python
df["mol"] = df.apply(lambda row: Chem.MolFromSmiles(row['SMILES']), axis=1)
df.head()
```

```python
from rdkit.ML.Descriptors import MoleculeDescriptors
calc = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
descriptors_list = [x[0] for x in Descriptors._descList]
print(descriptors_list)
```

```python
df['desc'] = df.apply(lambda row: calc.CalcDescriptors(row['mol']), axis=1)
header = descriptors_list
df_2D = pd.DataFrame(df['desc'].tolist(), columns = header)
df_2D_desc = pd.concat([df_InChIKey, df_2D], axis = 1)
```
